```python
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('covid.csv')
```

# Data1

```python
data = pd.DataFrame(data)
```

```python
values = np.array(data['Daily_Confirmed'].tolist())
plt.figure(figsize=(15,12))
plt.plot(values,color='red',marker='.',label='time series')
plt.ylabel('covid data',fontsize=30)
plt.title('Time Series',fontsize=40)
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(fontsize=20,facecolor='skyblue',edgecolor='black',framealpha=.5,loc='upper left')
plt.savefig('TS1.pdf')
def func(a,b):
    mean = np.mean(values[a:b])
    variance = np.var(values[a:b])
    return mean , variance
print(func(20,40))
for i in range(0,len(values),20):
    k = func(i,i+20)
    print("The mean and variance of",i, "to",i+20,"elements is",k)

print()
print("In this series we can seen that mean of elements is varying faster so this is not a stationary time series.")
    
```

# Converting in stationary time series

# 1. first difference

```python
# To make series stationary we will taking difference

value_diff = []
for i in range(len(values)-1):
    k = values[i+1]-values[i]
    value_diff.append(k)

plt.figure(figsize=(14,14))
plt.plot(value_diff,linestyle='dashed',color='green')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.savefig('TS2.pdf')

def func(a,b):
    mean = np.mean(value_diff[a:b])
    variance = np.var(value_diff[a:b])
    return mean , variance
#print(func(20,40))
for i in range(0,len(value_diff),20):
    k = func(i,i+20)
    print("The mean and varince of",i, "to",i+20,"elements is",k)
    
print()
print("In this series we can seen that mean of elements is not constant so this is not a stationary time series.")
    
```

#   2.  second difference

```python
# To make series stationary we will taking second difference

value_diff1 = []
for i in range(len(value_diff)-1):
    k = value_diff[i+1]-value_diff[i]
    value_diff1.append(k)

plt.figure(figsize=(14,14))
plt.plot(value_diff1,linestyle='dashed',color='green')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.ylabel('Data',fontsize=30)
plt.savefig('TS3.pdf')
def func(a,b):
    mean = np.mean(value_diff1[a:b])
    variance = np.var(value_diff1[a:b])
    return mean , variance
#print(func(20,40))
for i in range(0,len(value_diff1),20):
    k = func(i,i+20)
    print("The mean and varince of",i, "to",i+20,"elements is",k)

print()
print()
print("' since after taking second difference we can see that mean of series is not vary faster as we seen in previous case." )
print("so we can says this is stationary time series")
```
